```
import glob
import os
import cv2
import face_recognition

# Define the path to the input video files
input_video_files = glob.glob('/content/drive/My Drive/dataset_unzip/Celeb_fake_face/*.mp4')

# Define the output directory
output_directory = '/content/drive/My Drive/Pre_processed/pre_Celeb_fake_face/'

# Ensure the output directory exists
os.makedirs(output_directory, exist_ok=True)

# Function to extract frames from a video
def frame_extract(path):
    vidObj = cv2.VideoCapture(path)
    success = 1
    while success:
        success, image = vidObj.read()
        if success:
            yield image

# Function to create face videos
def create_face_videos(path_list, out_dir):
    print("Number of video files found:", len(path_list))
    for path in path_list:
        out_path = os.path.join(out_dir, os.path.basename(path))
        if os.path.exists(out_path):
            print("File Already exists: ", out_path)
            continue
        print("Processing video:", path)
        frames = []
        out = cv2.VideoWriter(out_path, cv2.VideoWriter_fourcc('M', 'J', 'P', 'G'), 30, (112, 112))
        for idx, frame in enumerate(frame_extract(path)):
            if idx <= 150:
                frames.append(frame)
                if len(frames) == 4:
                    faces = face_recognition.batch_face_locations(frames)
                    for i, face in enumerate(faces):
                        if len(face) != 0:
                            top, right, bottom, left = face[0]
                            try:
                                out.write(cv2.resize(frames[i][top:bottom, left:right, :], (112, 112)))
                            except Exception as e:
                                print(f"Error writing frame to video: {e}")
                    frames = []
        out.release()

# Calling the function with the appropriate paths
create_face_videos(input_video_files, output_directory)
```
